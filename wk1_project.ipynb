Importing necessary Libraries

```python
import pandas as pd

```

Load the Diabetes CSV dataset

```python
# Read the Diabetes database file into a dataframe
diabetes_database = pd.read_csv("/content/diabetes database.csv")
```

```python
# head()
diabetes_database.head()
```

```python
# info()
diabetes_database.info()
```

```python
# describe()
diabetes_database.describe()
```

```python
# How many patients have diabetes?
diabetes_count = diabetes_database['Outcome'].sum()

print("Patients with diabetes:", diabetes_count)
```

The result makes sense because the Outcome column uses 1 for patients with diabetes and 0 for patients without diabetes.

```python
# What is the average glucose level?
average_glucose = diabetes_database['Glucose'].mean()

print("Average glucose level:", average_glucose)
```

The result makes sense because the mean() function calculates the average glucose level for all patients in the dataset.

```python
# Show patients above age 40 (Gender column not found)
female_above_40 = diabetes_database[
    (diabetes_database['Age'] > 40)
]

female_above_40
```

The result makes sense because the filter selects only female patients whose age is greater than 40.

```python
# Group by outcome and show average BMI
average_bmi = diabetes_database.groupby('Outcome')['BMI'].mean()

print(average_bmi)
```

The result makes sense because grouping by Outcome separates diabetic and non-diabetic patients before calculating the average BMI for each group.

```python
# Drop rows with missing or zero glucose values
original_rows = len(diabetes_database)

clean_diabetes_database = diabetes_database[diabetes_database['Glucose'] != 0]

new_rows = len(clean_diabetes_database)

rows_removed = original_rows - new_rows

print("Rows removed:", rows_removed)
```

The result makes sense because rows with glucose value 0 are unrealistic and were removed to improve data quality.

TASK 3 — Five Chart Types

Import visualization libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Histogram
plt.hist(diabetes_database['Glucose'])

plt.title("Distribution of Glucose Levels")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
```

```python
# Scatter Plot
plt.scatter(diabetes_database['Age'], diabetes_database['Glucose'])

plt.title("Age vs Glucose")
plt.xlabel("Age")
plt.ylabel("Glucose")

plt.show()
```

```python
# Bar Chart
outcome_counts = diabetes_database['Outcome'].value_counts()

outcome_counts.plot(kind='bar')

plt.title("Diabetes Outcome Count")
plt.xlabel("Outcome")
plt.ylabel("Count")

plt.show()
```

```python
# Box Plot
plt.boxplot(diabetes_database['BMI'])

plt.title("BMI Distribution")
plt.ylabel("BMI")

plt.show()
```

```python
# Line Chart
age_glucose = diabetes_database.groupby('Age')['Glucose'].mean()

plt.plot(age_glucose.index, age_glucose.values)

plt.title("Average Glucose by Age")
plt.xlabel("Age")
plt.ylabel("Average Glucose")

plt.show()
```

Enable seaborn style

```python
# Enable seaborn style
sns.set_style("whitegrid")
```

```python
# Seaborn Histogram
sns.histplot(diabetes_database['Glucose'])

plt.title("Distribution of Glucose Levels")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")

plt.show()
```

```python
# Seaborn Scatter Plot
sns.scatterplot(x='Age', y='Glucose', data=diabetes_database)

plt.title("Age vs Glucose")
plt.xlabel("Age")
plt.ylabel("Glucose")

plt.show()
```

```python
# Seaborn Bar Chart
sns.countplot(x='Outcome', data=diabetes_database)

plt.title("Diabetes Outcome Count")
plt.xlabel("Outcome")
plt.ylabel("Count")

plt.show()
```

```python
# Seaborn Box Plot
sns.boxplot(y=diabetes_database['BMI'])

plt.title("BMI Distribution")
plt.ylabel("BMI")

plt.show()
```

```python
# Seaborn Line Chart
sns.lineplot(x=age_glucose.index, y=age_glucose.values)

plt.title("Average Glucose by Age")
plt.xlabel("Age")
plt.ylabel("Average Glucose")

plt.show()
```

I prefer the seaborn charts because they have cleaner styling, better readability, and a more professional appearance than the default matplotlib charts.

TASK 4 — Data Story

1. Does glucose level affect diabetes outcome?
2. Is BMI linked to diabetes outcome?
3. Which age group has the highest diabetes rate?

```python
# Glucose vs Outcome
sns.boxplot(x='Outcome', y='Glucose', data=diabetes_database)

plt.title("Glucose Level by Diabetes Outcome")
plt.xlabel("Outcome")
plt.ylabel("Glucose")

plt.show()
```

Patients with diabetes had a higher average glucose level than patients without diabetes.

```python
# BMI vs Outcome
sns.boxplot(x='Outcome', y='BMI', data=diabetes_database)

plt.title("BMI by Diabetes Outcome")
plt.xlabel("Outcome")
plt.ylabel("BMI")

plt.show()
```

Patients with diabetes generally had higher BMI values compared to non-diabetic patients.

```python
# Age Group with Highest Diabetes Rate
sns.histplot(data=diabetes_database, x='Age', hue='Outcome', multiple='stack')

plt.title("Age Distribution by Diabetes Outcome")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()
```

Middle-aged patients showed a higher number of diabetes cases compared to younger patients.
